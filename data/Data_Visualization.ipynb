## wp_yaw_const.scv
This file is used by styx.launch

```python
# This file is used by styx.lanch
#./waypoint_loader/launch/waypoint_loader.launch:4:        
# <param name="path" value="$(find styx)../../../data/wp_yaw_const.csv" />
with open ('./wp_yaw_const.csv') as f:
    wp_yaw_const_df = pd.read_csv(f, header=None, names=['x','y','z','yaw'], lineterminator='\n')
    plt.figure()
    plt.plot(wp_yaw_const_df.x,wp_yaw_const_df.y,linewidth=1,color='r')
    def arrow(table, index, plt, k=100):
        psi = table.iloc[index].yaw
        x = table.iloc[index].x
        y = table.iloc[index].y
        dx = math.cos(psi)*k
        dy = math.sin(psi)*k
        plt.arrow(x,y,dx,dy,fc="k", ec="k",linewidth=2,head_width=k/4, head_length=k/4)
    
    arrow(wp_yaw_const_df,5800,plt)
    arrow(wp_yaw_const_df,10900,plt)
    print('count:')
    print(wp_yaw_const_df.count())
```

## churchlot_with_cars.csv
This file used by styx.launch

```python
# This file used by styx.launch
#./waypoint_loader/launch/waypoint_loader_site.launch:4:        
# <param name="path" value="$(find styx)../../../data/churchlot_with_cars.csv" />
with open ('./churchlot_with_cars.csv') as f:
    churchlot_df = pd.read_csv(f, header=None, names=['x','y','z','yaw', 'unknown'], lineterminator='\n')
    plt.figure()
    plt.plot(churchlot_df.x,churchlot_df.y,linewidth=1,color="r")
    arrow(churchlot_df,20,plt,2)
    arrow(churchlot_df,50,plt,2)
    print('count:')
    print(churchlot_df.count())
```
